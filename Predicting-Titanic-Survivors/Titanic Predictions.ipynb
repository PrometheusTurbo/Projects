# Predicting Titanic Survival

Kaggle competition I registered for. This'll be my first Kaggle competition I'm competiting in! Let's talk a bit more about the data:

### train.csv



This contains all the details of the passengers on board. The first column is the passenger ID (1 - 891).

The second column (Survived) is telling us whether each person survived the crash or not.
- 1 if the person survived
- 0 if the person died

Pclass is a column that represents your ticket class.
- 1 is the first class (best)
- 2 is the second class
- 3 is the third class (worst)

SibSp is the number of siblings/spouse(s) on the Titanic.

parch is the number of parents/children(s) on the Titanic. 

And the rest of the columns are pretty self explanatory.

### test.csv

This is what we'll be testing on our model to see whether the other 418 passengers on board survived. Remember that there isn't a Survived column like train.csv. That's because Survived is what we want to prefict.

## Importing Libraries

Ok, enough talk about the data. Now let's get coding!

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
train_data = pd.read_csv("data/train.csv")
test_data = pd.read_csv("data/test.csv")
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
y = train_data["Survived"]
```

```python
features = ["Pclass", "Sex", "SibSp", "Parch"]
```

```python
x = pd.get_dummies(train_data[features]) #get_dummies is a Pandas function that helps make categorical values into raw values
x_test = pd.get_dummies(test_data[features]) 
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
ran_model = RandomForestClassifier(random_state=1, max_depth=32)
ran_model.fit(x,y)
ran_pred = ran_model.predict(x_test)
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': ran_pred})
output.to_csv('final submissions/ran_submission.csv', index=False)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

l_model = LinearRegression().fit(x,y)
l_pred = l_model.predict(x_test)

def round_linear(pred):
    if (1 - pred) < (pred - 0):
        return 1
    else: 
        return 0

pred_list = []
    
for pred in l_pred:
    pred_list.append(round_linear(pred))
    
l_pred = np.asarray(pred_list)
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': l_pred})
output.to_csv('final submissions/l_submission.csv', index=False)
```

## K-Nearest Neighbour (KNN)

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline 

distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(x)
    distortions.append(kmeanModel.inertia_)
    
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=4) #4 is the optimal k value
knn_model.fit(x,y)
knn_pred = knn_model.predict(x_test)
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': knn_pred})
output.to_csv('final submissions/knn_submission.csv', index=False)
```
